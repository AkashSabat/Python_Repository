```python
# We will be crawling the HackerNews website.
# HackerNews is a website where one can read about latest programming news and Industry Innovations.
# Here we are using beautiful Soup to extract data from websites.
# Beautiful Soup allows us to use html and grab different data.
```

```python
# importing requests and beautiful soup
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://news.ycombinator.com/news')
print(res.text)
```

```python
# Parse the Html using BeautifulSoup
soup = BeautifulSoup(res.text,'html.parser')
print(soup)
```

```python
# printing the entire body
print(soup.body)
```

```python
#printing the entire Contents
print(soup.body.contents)
```

```python
# using the findAll() method to find all the divs in the soup object
print(soup.find_all('div'))
```

```python
# findAll() method to find all the a tags i.e links
print(soup.find_all('a'))
```

```python
# extract all the links
links = soup.select('.storylink')
print(links)
```

```python
votes = soup.select('.score')
print(votes[0])
```

```python
# Get all the links using function

def create_custom_hn(links,votes):
    hn = []
    for idx, item in enumerate(links):
        title = links[idx].getText()
        href = links[idx].get('href',None)
        hn.append(href)
        return hn
        
print(create_custom_hn(links,votes))
```

```python
#  extract  the title and the links

def create_custom_hn(links,votes):
    hn = []
    for idx, item in enumerate(links):
        title = links[idx].getText()
        href = links[idx].get('href',None)
        hn.append({'title':title, 'link':href})
        return hn
        
print(create_custom_hn(links,votes))
```

```python
# extract the votes

def create_custom_hn(links,votes):
    hn = []
    for idx, item in enumerate(links):
        title = links[idx].getText()
        href = links[idx].get('href',None)
        points = int(votes[idx].getText().replace(' points',''))
        print(points)
        hn.append({'title':title, 'link':href})
        return hn
        
print(create_custom_hn(links,votes))
```
